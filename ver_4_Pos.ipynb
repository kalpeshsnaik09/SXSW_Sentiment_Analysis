```python
sp=spacy.load('en_core_web_lg')
sp_stopwords = sp.Defaults.stop_words
lemmatizer = WordNetLemmatizer()
```

```python
#Create a function to clean tweets
def cleanText(text):
    text=str(text)  #Coverts Text to String
    text=re.sub(r'@[A-Za-z0-9]+','',text)  #Removing @Mentions
    text = re.sub(r'#[\w]*sxsw[\w]*', ' ', text,flags=re.I)  #Removing sxsw hashtag
    text=re.sub(r'#','',text)  #Removing # Symbols
    text=re.sub(r'RT[\s]+','',text)  #Removing ReTweets
    text=re.sub(r'https?:\/\/\s+','',text)  #Removing the hyperlinks
    text=text.replace(r'{html}',"") 
    cleanr = re.compile(r'<.*?>')
    text = re.sub(cleanr, '', text)
    text = re.sub(r'[0-9]+', '', text)  #Removing Numbers
    text = re.sub(r'[^A-Za-z]+', ' ', text)  #Removing all spacial character
    text = text.lower()  #Coverts Text To Lower Case
    
#     doc=sp(text)
#     return ' '.join([token.lemma_ for token in doc if token not in stopwords and len(token)>2])
    
    tokenizer = RegexpTokenizer(r'\w+')
    tokens = tokenizer.tokenize(text)  
#     filtered_words = [w for w in tokens if len(w) > 2 and w not in stopwords.words('english')]
#     stem_words=[stemmer.stem(w) for w in filtered_words]
    lemma_words=[lemmatizer.lemmatize(w) for w in tokens if w not in ['sxsw','amp','quot','http','html']]
    return " ".join(lemma_words)
```

```python
data['clean_text']=data['tweet'].map(lambda s:cleanText(s))
```

```python
for i in data['clean_text']:
    for ent in sp(str(i)).ents:
        if ent.label_=='ORG':
            org.append(ent.text)
        elif ent.label_=='PRODUCT':
            product.append(ent.text)
```
